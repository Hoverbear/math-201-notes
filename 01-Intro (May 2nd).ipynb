## Object of Study
A differential equation relates an unknown function and one of more derivatives.

### Ex
$$ x: \mathbb{R} \rightarrow \mathbb{R} $$
$$ x = x(t) $$
$$ x\prime(t) = x(t) $$
One solution: $ x_1(t)=e^t $

Another Solution $ x_2(t)=0 $

## Directions of Study

* **Exact Methods**: Given an equation, find all solutions. (Very hard to do)
* **Numerical Methods**: Use computers (today!) to obtain approximations of specific solutions.
* **Qualitative Methods**: Obtain properties of the solutions without knowing the solution explicitly.

## Classification

* **Domain of the Unkown Function**
    * Ordinary differential equation (**ODE**): Depends on a single variable.
    * Partial differential equation (**PDE**): Depend on more then one variable.
* **Number of Unknowns**
    * For example:
        $$ x = x(t), y=y(t) $$
        $$ x\prime = 2y $$
        $$ y\prime = x+3y $$
        One solution: x(t)=y(t)=0
* **Structure of the Equation**
    * Linear: The unknowns are functions at power 1
    * Nonlinear: The unknowns are functions that are not linear.
* **Order of the Equation**
    * Order $k$ if the highest derivative is the $k^{th}$ derivative.
    * Ex. Second order: $ x\prime\prime = 2x $

 ## Applications

* Growth and Decay Problems
    $$ x\prime = k*x $$
    $$ k = const $$
    * Determines the half-life of radioactive substances
    * Doubling of investments.
* Newton's law of cooling and heating.
    $$ T\prime = k(t-\tau) $$
    $$ k,\tau = Const, T=Temperature $$
* Chemistry Mixing Problems
    $$ u\prime = k(a-u)(b-u) $$
    $$ k, a, b = const $$
* Logistic Equation
    $$ p\prime = K p(\alpha-p) $$
    $$ a, K = const $$
* Electric Circuits and Springs with period forcing tension
    $$ x\prime\prime + bx\prime + kx = \gamma \sin{\omega t} $$
    $$ b, k, \gamma, omega = const $$
* Lorenz Equation
    $$ x\prime = \sigma(y-x) $$
    $$ y\prime = \rho x-y-xz $$
    $$ z\prime = \beta z + xy $$
    * Used to model compressed fluid/gas
* 3-Body Problem
    $$ q_{1,i}\prime\prime = Gm_2\frac{q_{2,i}-q_{1,i}}{r_{12}^3}+Gm_2\frac{q_{3,i}-q_{1,i}}{r_{21}^3} $$
    $$ q_{2,i}\prime\prime = Gm_1\frac{q_{1,i}-q_{2,i}}{r_{12}^3}+Gm_3\frac{q_{3,i}-q_{2,i}}{r_{32}^3} $$
    $$ q_{3,i}\prime\prime = Gm_1\frac{q_{1,i}-q_{3,1}}{r_{13}^3}+Gm_3\frac{q_{2,i}-q_{3,i}}{r_{23}^3} $$
    * $G=$ Gravitiational Constant
    * $q_{j,i} =$ The $i^{th}$ of the $j^{th}$ body in a 3-dimensional frame.
    * $r_{j,i} =$ Distance between the $i^{th}$ and $j^{th}$ body.

### Two Body Problem
Force: $ \frac{GmM}{x^2+y^2} $

$m, M =$ Masses

$G=$ Graviational Constants

$$ u = \frac{(-x,-y)}{(x^2+y^2)^{1/2}} $$
$$ F = \frac{(-x,-y)GgM}{(x^2+y^2)^{3/2}} $$

$$ F = ma $$
$$ F_x = Mx\prime\prime $$
$$ F_y = My\prime\prime $$

It follows that:
$$ x\prime\prime = -\frac{Gmx}{(x^2+y^2)^{3/2}} $$
$$ y\prime\prime = -\frac{Gmy}{(x^2+y^2)^{3/2}} $$
