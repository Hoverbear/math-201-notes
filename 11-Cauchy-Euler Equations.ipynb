## Problem Of the Form
$$ a_nx^n \frac{d^ny}{dx^n} + a_{n-1}x^{n-1} \frac{d^{n-1}}{dx^{n-1}}+...+a_1x\frac{dy}{dx}+a_0y=g(x) $$
Note the non-constant coefficients (depending on x)

The idea is to **always check the form** $y(x)=x^m$ in the homogenious equation.
To solve the non-homogenious you can use variation of parameters or undetermined coefficients.

**We must assume that x is not zero, so we restrict ourselves from 0 to infinity.**

## Second Order Case
$$ ax^2y\prime\prime + bxy\prime + cy = g(x) $$

We're first interested in the homogenious equation:
$$ ax^2y\prime\prime + bxy\prime + cy = 0 $$
(Once this is solved we'll use variation of parameters or undetermined coefficients.)

$$ y(x) = x^m $$
$$ y\prime(x) = mx^{m-1} $$
$$ y\prime\prime(x)= m(m-1)x^{m-2} $$

Bringing this into the equation:
$$ ax^2m(m-1)x^{m-2}+bxmx^{m-1}+cx^m = 0 $$
$$ am(m-1)x^m + bmx^m + cx^m = 0 $$

Dividing by $x^m, x\in(0,\infty)$:
$$ am(m-1)+bm+c=0 $$
$$ am^2+(b-a)m+c = 0 $$

There are **three cases**:

### Distinct Real Roots
$m_1, m_2$ roots, $m_1\neq m_2$.
$$ y_1(x)=x^{m_1}, y_2(x)=x^{m_2} $$

This is linearly independant.
So the general solution of the homogenious equation:
$$ y_c(x)=c_1x^{m_1}+c_2x^{m_2} $$

**Ex.**
$$ x^2y\prime\prime-2xy\prime-4y=0 $$

$$ y(x)=x^m $$
$$ y\prime(x)=mx^{m-1} $$
$$ y\prime\prime(x)=m(m-1)x^{m-2} $$

$$ x^2m(m-1)x^{m-2}-2xmx^{m-1}-4x^m = 0 $$
$$ (m^2-3m-4)x^m = 0 $$
Dividing by $x^m$:
$$ (m^2-3m-4)=0 $$
Factoring:
$$ m_1= -1, m_2=4 $$

Therefore:
$$ y(x)=c_1x^{-1}+c_2x^4 $$

### Real, Repeated Roots
$$ m_1=m_2 \in \mathbb{R} $$
$$ y_1{x}=x^{m_1} $$
(From the quadratic equation)
$$ m_1 = \frac{a-b}{2a} $$
To find a second linearly independant solution, use reduction (4.2)... in other words check a solution of the form:
$$ y_2(x)=u(x)x^{m_1} $$

After dividing by $ax^2$:
$$ y\prime\prime+\frac{b}{ax}y\prime + \frac{c}{ax^2}y = 0 $$

*Recall*: $ y\prime\prime+ P(x)y\prime + Q(x)y = 0 $
$$ y_2(x)=y_1(x)\int{\frac{e^{-\int{P(x)dx}}}{y_1^2(x)}}dx $$

So, $P(x)=\frac{b}{ax}$
$$ \int{P(X)dx}=\frac{b}{a}\ln{|x|} $$
$$ y_2(x)=x^{m_1}\int{\frac{e^{-\frac{b}{a}\ln{|x|}}}{x^{2m_1}}}dx $$
Simplifying:
$$ y_2(x)=x^{m_1}\int{x^{-\frac{b}{a}}x^{-2m_1}}dx $$
$$ y_2(x)= x^{m_1}=\int{x^{-\frac{b}{a}}x^{\frac{b-a}{a}}}dx $$
$$ y_2(x)=x^{m_1}=\int{\frac{dx}{x}}=x^{m_1}\ln{x} $$General Solution:
$$ y(x)=c_1x^{m_1}+c_2x^{m_1}\ln{|x|} $$

**Ex.**
$$ 4x^2y\prime\prime + 8xy\prime + y = 0 $$
$$ y(x)=x^m $$
$$ y\prime(x)=mx^{m-1} $$
$$ y\prime\prime(x)=m(m-1)x^{m-2} $$
$$ x^m(4m^2+4m+1)=0 $$
$$ m_1=m_2=-\frac{1}{2} $$
$$ y(x)=c_1x^{-\frac{1}{2}}+c_2x^{-\frac{1}{2}} \ln{x} $$

For higher orders we obtain linearly independent solutions:
$$ x^{m_1}, x^{m_1}\ln{x}, x^{m_1}(\ln{x})^2, ... , x^{m_1}(\ln{x})^{k-1} $$

### Complex Conjugate Roots
$$ m_1= \alpha+i\beta $$
$$ m_2 = \alpha-i\beta $$
$$ \alpha, \beta \in \mathbb{R}, \beta > 0 $$

$$ y(x)=c_1x^{\alpha+i\beta} + c_2x^{\alpha-i\beta} $$
But these are complex. To obtain real solutions,
$$ x^{i\beta} = (e^{\ln{x}})^{i\beta} = e^{i\beta\ln{x}} $$

Using Eulers formula:
$$ x^{i\beta} = \cos{(\beta\ln{x})}+i\sin{(\beta\ln{x})} $$
Similarily:
$$ x^{-i\beta} = \cos{(\beta\ln{x})}-i\sin{(\beta\ln{x})} $$
Noting:
$$ x^{i\beta} + x^{-i\beta} = 2\cos{(\beta\ln{x})} $$
$$ x^{i\beta}-x^{-i\beta} = 2i\sin{(\beta\ln{x})} $$

Take $ c_1=c_2=1 $:
$$ y_1=x^\alpha(x^{i\beta}+x^{i\beta}) $$

Take $$ c_1=1, c_2 = -1 $$
$$ y_2=x^\alpha(x^{i\beta}-x^{-i\beta})=2ix^\alpha \sin{(\beta \ln{x})} $$

Checking for independence:
$$ W(x^\alpha\cos{(\beta\ln{x})}, x^\alpha\sin{(\beta\ln{x})}) = \beta x^{2\alpha-1} \neq 0 $$

General Solution **memorize**:
$$ y(x)=c_1x^\alpha \cos{(\beta\ln{x})}+c_2x^\alpha \sin{(\beta \ln{x})} $$

**Ex.**
$$ 4x^2y\prime\prime + 17y = 0 $$
$$ y(1)=-1 $$
$$ y\prime(1) = - \frac{1}{2} $$

$$ y(x)=x^m $$
$$ 4x^2y\prime\prime + 17y = x^m (4m(m-1)+17) = 0 $$
$$ 4m^2-4m+17 = 0 $$
$$ m_1 = \frac{1}{2}+2i, m_2 = \frac{1}{2}-2i $$
$$ \alpha = \frac{1}{2}, \beta = 2 $$

From general solution:
$$ y(x)=x^{\frac{1}{2}}(c_1\cos{(2\ln{x})}+c_2\sin{(2\ln{x})} $$

But this is an IVP:
$$ -1 = y(1) = 1(c_1) $$

$$ y\prime(x)= \frac{1}{2}x^{-\frac{1}{2}}(c_1\cos{(2\ln{x})}+c_2\sin{(2\ln{x})} + x^{\frac{1}{2}}(-c_1\sin{(2\ln{x})}*\frac{2}{x} + c_2\cos{(2\ln{x})}*\frac{2}{x} $$
$$ -\frac{1}{2} = y\prime(1) = \frac{1}{2}(c_1+0) + 1(2c_2) $$
$$ c_2 = 0 $$

Solution of the IVP is:
$$ y(x)=-x^{\frac{1}{2}} \cos{(2\ln{x})} $$

### Third Order Example
$$ x^3y\prime\prime\prime + 5x^2y\prime\prime + 7xy\prime+ 8y = 0 $$
$$ y(x)=x^m $$
$$ y\prime(x) = mx^{m-1} $$
$$ y\prime\prime(x) = m(m-1)x^{m-2} $$
$$ y\prime\prime\prime(x) = m(m-1)(m-2)x^{m-3} $$

$$ x^m(m^3+2m^2+4m+8) = 0 $$
(This is easy to factor, just group the first 2 terms and the last two)
$$ m^2(m+2)+4(m+2) = 0 $$
$$ (m+2)(m^2+4) = 0 $$

$$ m_1 = -2, m_2 = 2i, m_3 = -2i $$

So the general solution is:
$$ y(x) = c_1x^{-2}+c_2\cos{(2\ln{x})}+c_3\sin{(2\ln{x})} $$

## Example
$$ x^2y\prime\prime + 3xy\prime + 3y = 2x^4e^x $$
$$ x^2y\prime\prime  - 3xy\prime + 3y = 0 $$

$$ y(x)=x^m $$

$$ m(m-1)x^m-3mx^m + 3x^m = 0 $$
$$ m^2-m-3m+3=0 $$
$$ m^2 -3m + 3 = 0 $$
Gives us roots:
$$ m_1,2 = \frac{4\pm\sqrt{16-12}}{2}=\frac{4\pm2}{2} = 3, 1 $$

$$ y_c(x)=c_1x+c_2x^3 $$

Now we use variation of parameters to get $y_p$:
$$ y_p(x) = u_1(x)y_1(x) + u_2(x)y_2(x) $$
Where $y_1=x, y_2=x^3$

$$ u_1\prime = \frac{W_1}{W}, u_2\prime = \frac{W_2}{W} $$
$$ y\prime\prime - \frac{3}{x}y\prime + \frac{3}{x^2}y = 2x^2e^x $$

$$ f(X) = 2x^2e^x $$
$$ W = \begin{vmatrix} x & x^3 \\\ 1 & 3x^2 \end{vmatrix} $$
$$ W_1 = \begin{vmatrix} 0 & x^3 \\\ 2x^2e^x & 3x^2 \end{vmatrix} $$
$$ W_2 = \begin{vmatrix} x & 0 \\\ 1 & 2x^2e^x \end{vmatrix} $$

$$ u_1\prime = -x^2 e^x \Rightarrow Integrate by parts twice \Rightarrow u_1(x)=-x^2e^x+2xe^x-2e^x $$
$$ u_2\prime = e^ \Rightarrow u_2(x)=e^x $$

$$ y_p(x)=(-2x^2e^x+2xe^x-2e^x)x+e^xx^3 = 2x^2e^x-2xe^x $$

Finally:
$$ y(x) =  y_c(x)+y_p(x)=c_1x+c_2x^3 + 2x^2e^x - 2xe^x $$

## Notice Simliarity?
$$ ay\prime\prime = by\prime + cy = 0 $$
$$ y(x)=c_1e^{m_1}+c_2e^{m_2x} $$

$$ 2x^2y\prime\prime + bxy\prime + cy = 0 $$
$$ y(x)=c_1x^{m_1}+c_2x^{m_2} $$

$$ e^{\ln{x}} = x, x > 0 $$
$$ y(x) = c_1e^{m_1\ln{x}} + c_2e^{m_2\ln{x}} + c_1e^{m_1t}+c_2e^{m_2t} $$

**Ex.**
$$ x^2y\prime\prime - xy\prime + y = 0, x>0 $$
$$ x=e^t, t=\ln{x} $$
(Chain rule fun!)

$$ \frac{dy}{dx} = \frac{dy}{dt}\frac{dt}{dx} = \frac{1}{x}\frac{dy}{dt} $$
$$ \frac{d^2y}{dx^2} = \frac{1}{x}\frac{d}{dx}(\frac{dy}{dt}) + (-\frac{1}{x^2})\frac{dy}{dt} $$

What is $ \frac{d}{dx}\frac{dy}{dt} $?
$$ \frac{d}{dx}\frac{dy}{dt} = \frac{d}{dx}y\prime(t(x)) = \frac{dy\prime}{dx}\frac{1}{x}=\frac{d^2y}{dt^2} $$

So, $$ \frac{d^2y}{dx^2} = \frac{1}{x^2}(\frac{d^2y}{dt^2}-\frac{dy}{dt}) $$

So we need to sub:
$$ x^2y\prime\prime - xy\prime + y = x^2 \frac{1}{x^2}(\frac{d^2y}{dt^2}-\frac{dy}{dy}) - x\frac{1}{x}\frac{dy}{dt}t + y = 0 $$
$$ \frac{d^2y}{dt^2}-2\frac{dy}{dt} + y = 0 $$

Hey look! A second order equation with constnat coefficients. We can **solve this!**
$$ m^2-2m+1 = 0 $$
$$ m_1=m_2=1 $$
$$ y_c(t)=c_1e^+c_2te^t $$

Do the substitution for $t \rightarrow x$:
$$ y_c(x)=c_1e^{\ln{x}} + c_2\ln{x}e^{\ln{x}} $$
$$ y_c(x) = c_1x+c_2x\ln{x} $$
