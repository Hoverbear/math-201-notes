Recall:
$$ ay\prime + by = 0 $$
We solved it as:

* Seperable
* Linear

$$ y\prime=-\frac{b}{a}y $$
Let $-\frac{b}{a} = k$
$$ y\prime = ky $$

Solution:
$$ ce^{mx} $$

Assume:
$$ y=e^{mx}, y\prime=me^{mx} $$

Substitute into:
$$ ay\prime + by = 0 $$
$$ ame^{mx} + be^{mx} = 0 $$
$$ e^{mx}(am+b) = 0 $$

Noting that $e^{mx}$ is always greater then 0.
$$ am+b = 0$$
$$ m = -\frac{b}{a} $$

## Can we use this idea to solve
$$ a_ny^n+a_{n-1}y^{n-1}+...+a_2y\prime\prime+a_1y\prime+a_0y = 0 $$
$ a_0, ..., a_n $ constant coefficients.

This is the **Characteristic Equation** or **Auxilary Equation**.

Start with 2nd order equation:
$$ ay\prime\prime + by\prime + cy = 0 $$
$$ a,b,c \in \mathbb{R}, a \neq 0 $$

Try to find solutions of the form $y(x)=e^{mx}$.
$$ y\prime(x) = me^{mx} $$
$$ y\prime\prime(x)=m^2e^{mx} $$

Substitute into the equation:
$$ am^2e^{mx}+bme^{mx}+ce^{mx} = 0 $$
$$ e^{mx}am^2+bm+c)=0 $$
Where $e^{mx} \neq 0$

So the characteristic equation is:
$$ am^2+bm+c=0 $$

## We have three possible situations

### Case 1: Real, distinct roots
$$ m_1, m_2 $$

$$ y_1(x)=e^{m_1x}, y_2{x}=e^{m_2x} $$
$$ y_1, y_2: \mathbb{R} \rightarrow \mathbb{R} $$

So check with the Wranskin
$$ W(y_1, y_2) = \begin{vmatrix} e^{m_1x} & e^{m_2x} \\\ m_1e^{m_1x} & m_2e^{m_2x} \end{vmatrix} = m_2e^{m_1x}e^{m_2x} - m_1e^{m_1x}e^{m_2x} = (m_2-m_1)e^{(m_1+m_2)x} \neq 0 $$

### Case 2: Real, repeated roots
$$ m_1 = m_2 =: m $$

We have one solution $y_1(x)=e^{mx}$ to find the second one, use reduction $y(x)=u(x)e^{mx}$

(homework)

Thus you find a second solution:
$$ y_2(x)=xe^{mx} $$
$ y_1,y_2$ linearly independant

(homework)

**General solution:**
$$ y(x) = c_1e^{mx}+c_2xe^{mx} $$
$$ c_1, c_2 \in \mathbb{R} $$

### Case 3: Complex Conjugate roots
$$ m_1 = \alpha+i\beta, m_2 = \alpha - i\beta $$

Proceeding as in **Case 1** we can obtain:
$$ y(x)=c_1e^{(\alpha-i\beta)x}+c_2e^{(\alpha-i\beta)x} $$
$$ c_1, c_2 \in \mathbb{R} $$

Using **Euler's Formula**:
$$ e^{i\theta} = \cos{\theta}+i\sin{\theta} $$
$$ e^{i\pi}+1=0 $$

$$ e^{i\beta x} = \cos{\beta x} + i\sin{\beta x} $$
$$ e^{-i\beta x} = \cos{\beta x} - i\sin{\beta x} $$

$$ e^{i\beta x} + e^{-i\beta x} = 2\cos{\beta x} $$
$$ e^{i\beta x} - e^{-i\beta x} = 2\sin{\beta x} $$

$$ y = c_1e^{(\alpha+i\beta)x} + c_2e^{(\alpha-i\beta)x} $$
$$ c_1 =c_2 = 1$$
$$ y_1 = e^{(\alpha+i\beta)x} + e^{(\alpha-i\beta)x} $$
$$ c_1 = 1, c_2 = -1 $$
$$ y_2=e^{(\alpha+i\beta)x}-e^{(\alpha-i\beta)x} $$

$$ y_1 = e^{\alpha x}(e^{i\beta x} + e^{-i\beta x})  = 2e^{\alpha x}\cos{\beta x} $$
$$ y_2 = e^{\alpha x}(e^{i\beta x}-e^{-i\beta x}) = 2ie^{\alpha x}\sin{\beta x} $$

Either use $y_1$ and reduction to obtain $y_3$ or notice that $y_3= 2e^{\alpha x}\sin{\beta x}$ is also a solution (real).

**General solution** (memorize):
$$ y(x)=e^{\alpha x}(c_1\cos{\beta x} + c_2\sin{\beta x}) $$
$$ c_1, c_2 \in \mathbb{R} $$

### Ex. Case 1
$$ 2y\prime\prime - 5y\prime - 3y= 0$$

Characteristic Equation:
$$ 2m^2 - 5m - 3 = 0 $$
$$ m_1 = -\frac{1}{2}, m_2 = 3 $$

General solution:
$$ y(x)=c_1e^{-\frac{x}{2}} + c_2e^{3x} $$
$$c_1, c_2  \in \mathbb{R} $$

## Nice Equations with Memorable Solutions
### Ex.
$$ y\prime\prime + k^2y = 0 $$

Characteristic:
$$ m^2 + k^2 = 0 $$
$$ m_{1,2} = \pm ki $$
$$ \alpha = 0, \beta = k $$

General Solution:
$$ y(x)=c_1\cos{kx}+c_2\sin{kx} $$

### Ex.
$$ y\prime\prime - k^2y = 0 $$

Characteristic:
$$ m^2-k^2 = 0 $$
$$ m_1 = k, m_2 = -k $$

General SOlution:
$$ y(x)=c_1e^{kx}+c_2e^{-kx} $$

Also:
$$ c_1 = \frac{1}{2}, c_2 = -\frac{1}{2} $$
$$ y_1=\frac{1}{2}(e^{\frac{1}{2}x}-e^{-kx}) = \sinh{kx} $$

or

$$ c_1 = \frac{1}{2}, c_2 = -\frac{1}{2} $$
$$ y_2 = \frac{1}{2}(e^{kx}+e^{-kx}) = \cosh{kx} $$

Are also independant solutions.

$$ y(x) = c_1\cosh{kx} + c_2 \sinh{kx} $$
