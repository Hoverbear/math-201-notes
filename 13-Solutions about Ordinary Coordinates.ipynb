## Theorem
If $x=x_0$ is an ordinary point of a differential equation with polynomial coefficients, we can always fine two linearly independent solutions in the form of a power series centered at $x_0$, that is
$$ y(x) = \sum_{n=0}^{\infty}{ c_n(x-x_n)^n } $$
A power series solution converges at least on some interval $|x-x_0| \lt R$, where $R$ is the distance from $x_0$ to the closest singular point.

**Ex.**

$$ (x^2+1)y\prime\prime + xy\prime + x^3y = 0 $$
If we seek a power series solution about $x=0$, it will be convergent in the interval $(-1,1)$ at least. (From theorem)

An example where the coefficients are not polynomials

$$ y\prime\prime+(\cos{x})y=0 $$
We can find a power series solution $y(x)=\sum_{n=0}^{\infty}{ c_nx^n }$ (about $x=0$)

$$ y\prime\prime+(\cos{x})y = \sum_{n=2}^{\infty}{ n(n-1)c_nx^{n-2} } + (1-\frac{x^2}{2!}+\frac{x^4}{4!}-\frac{x^6}{6!}+...)\sum_{n=0}^{\infty}{ c_nx^n } $$
$$ = 2c_2 + 6c_3x + 12c_4x^2 + 20c_5x^3 + ... + (1-\frac{x^2}{2!}+\frac{x^4}{4!}-...)(c_0 + c_1x + c_2x^2 + c_3x^3 + ... ) $$

Group terms by $x$:
$$ = c_0+2c_2 + (c_1+6c_3)x + (-\frac{1}{2}c_0+c_2+12c_4)x^2 + (-\frac{1}{2}c_1+c_3+20c_5)x^3 + ... = 0 $$

So now we can at least compute the first few coefficients of the equation.
$$ \Rightarrow c_0 + 2c_2 = 0 $$
$$ \Rightarrow c_1+ 6c_3 = 0 $$
$$ \Rightarrow -\frac{1}{2}c_0 + c_2 + 12c_4 = 0 $$
$$ \Rightarrow -\frac{1}{2}c_1 + c_3 + 20c_5 = 0 $$

$$ c_2 = -\frac{1}{2}c_0 $$
$$ c_3 = -\frac{1}{6}c_1 $$
$$ c_4 = \frac{1}{12}c_0 $$
$$ c_3 = \frac{1}{30}c_1 $$

$$ y(x) = c_0y_1(x) + c_2y_2(x) $$
