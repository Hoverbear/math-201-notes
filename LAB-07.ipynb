# Variation of Parameters
Used for solving non-homogenious parts.

* First we solve the Homogenious.
* Then we use this method to find the particular solution.

## The Routine
$$ y\prime\prime+P(x)y\prime+Q(x)y=f(x) $$

Solve the **homogenious case:**
$$ y\prime\prime + P(x)y\prime +Q(x)y = 0 $$
To get $ y_c = c_1y_1+c_2y_2 $

Assume a solution of the form:
$$ y= u_1(x)y_1(x)+u_2(x)y_2(x) $$
$$ u_1\prime = \frac{W_1}{W}, u_2=\frac{W_2}{W} $$

Where (Should know all these!)
$$ W = \begin{vmatrix} y_1 & y_2 \\\ y_1\prime & y_2\prime \end{vmatrix} = y_1y_2\prime-y_2y_1\prime$$
$$ W_1 = \begin{vmatrix} 0 & y_2 \\\ f(x) & y_2\prime \end{vmatrix} = y_2f(x) $$
$$ W_2 = \begin{vmatrix} y_1 & 0 \\\ y_1\prime & 0 \end{vmatrix} = y_1f(x) $$

### Ex.
Solve using variation of parameters:
$$ y\prime\prime + y = \sin{x} $$

Expect it to look like $ y = c_1\cos{x}+c_2\sin{x}+Ax\cos{x}+Bx\sin{x} $

Getting $y_c$:
$$ m^2 + 1 = 0 $$
$$ m = \pm \sqrt{-1} = \pm i $$

$$ y_c = c_1\sin{x} + c_2\cos{x} $$
$$ y_1 = \sin{x}, y_2 = \cos{x} $$

$$ W = \begin{vmatrix} \sin{x} & \cos{x} \\\ \cos{x} & -\sin{x} \end{vmatrix} = -\sin^2{x}-\cos^2{x} = -1 $$
$$ W_1 = \begin{vmatrix} 0 & \cos{x} \\\ \cos{x} & \sin{x} \end{vmatrix} = -\sin^2{x}-\cos^2{x} = -\sin{x}\cos{x} $$
$$ W_2 = \begin{vmatrix} \sin{x} & 0 \\\ \cos{x} & \sin{x} \end{vmatrix} = -\sin^2{x}-\cos^2{x} = \sin^2{x} $$

$$ u_1\prime = \frac{W_1}{W} = \sin{x}\cos{x} $$
$$ u_1 = \int{\sin{x}\cos{x}}dx $$
Int Subsitution:
Let $ v=\sin{x} $
$$ dv = \cos{x}dx $$

$$ u_1 = \int{v}dv = \frac{v^2}{2}+c = \frac{\sin^2{x}+c_1}{2} $$

$$ u_2\prime = -\sin^2{x} $$
$$ u_2 = -\int{\sin^2{x}}dx = -\frac{1}{2} \int{1-\cos{2x}}dx = -\frac{1}{2}(x-\frac{\sin{2x}}{2})+c = -\frac{1}{2}(x-\cos{x}\sin{x}) + C_2 $$

Therefore
$$ y = \frac{\sin^2{x}}{2}\sin{x}+C_1\sin{x} - \frac{1}{2}x\cos{x}+\frac{1}{2}\cos^2{x}\sin{x} + c_2\cos{x} $$
Which we can factor into:
$$ y = \frac{1}{2}\sin{x}(\sin^2{x}+\cos^2{x})+c_1\sin{x}-\frac{1}{2}x\cos{x}+c_2\cos{x} $$
Which leaves us with:
$$ y= \frac{1}{2}\sin{x}(1) + c_1\sin{x}+c_2\cos{x}-\frac{1}{2}x\cos{x} $$
$$ y = (\frac{1}{2}+c_1)\sin{x}+c_2\cos{x}-\frac{1}{2}x\cos{x} $$

**General Solution**:
$$ y = c_3\sin{x} + c_2\cos{x} - \frac{1}{2}x\cos{x} $$
