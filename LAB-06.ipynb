### Looking at the homogenious equation (to find $y_c$):
$$ y\prime\prime - 5y\prime + 4y = 0 $$

Using that we get
$$ m^2e^{mx}-5me^{mx}+4e^{mx} = 0 $$
Getting the roots:
$$ m^2-5m+4 = 0 $$
$$ m=1,4 $$

So our characteristic equation is:
$$ y_c=c_1e^x+c_2e^{4x} $$

Recall that $y = y_c + y_p$.

### Finding $y_p$:
If I had $F(x)$ so that
$$ F\prime\prime(x)-5F\prime(x)+4F(x)=x^2 $$
and $g(x)$ so that
$$ g\prime\prime(x)-5g\prime(x)+4g(x)=x\sin{x} $$

Then $y=F(x)+g(x)$ solves
$$ y\prime\prime-5y\prime+4y=x^2x\sin{x} $$

**Why is this justified?**
$$ (F(x)+g(x))\prime\prime - 5(F(x)+g(x))\prime+4(F(x)+g(x)) $$
$$ = F\prime\prime +g\prime\prime - 5(F\prime+g\prime)+4(F+g) $$
$$ = (F\prime\prime-5F\prime+4F)+(g\prime\prime-5g\prime+4g) $$
$$ = x^2 + x\sin{x} $$

So to solve this:

1. Find $F$, in other words, solve:
    $$ y\prime\prime-5y\prime+4y=x^2 $$
    Try:
    $$ y_{p1} = Ax^2+Bx+C $$
    $$ y_{p1}\prime\prime-5y_{p1}\prime+4y_{p1} = (2A)-5(2Ax+B)+4(Ax^2+Bx+C) $$
    Regroup according to powers of X.
    $$ 4Ax^2+(-10A+4B)x+(2A-5B+4C) = x^2 $$
    So we get equations:
    $$ 4A=1 \rightarrow A=\frac{1}{4} $$
    $$ -10A+4B=0 \rightarrow B=\frac{1}{4}\frac{5}{2} = \frac{5}{8} $$
    $$ 2A-5B+4C=0 \rightarrow C=\frac{1}{4}(\frac{25}{8}-\frac{1}{2})=\frac{21}{32} $$
    Therefore:
    $$ y_{p1}=\frac{1}{4}x^2+\frac{5}{8}x+\frac{21}{32} $$
2. Find $g$, in other words, solve:
    $$ y\prime\prime-5y\prime+4y=x\sin{x} $$
    Try:
    $$ (Ax+B)\cos{x}+(Cx+D)\sin{x} $$
    **Start: Why?**
    $$ \frac{d}{dx}(x\sin{x}) = \sin{x}+x\cos{x} $$
    $$ \frac{d}{dx}(x\cos{x})=\cos{x}-x\sin{x} $$
    **End: Why?**

    $$ y_{p2} = (Ax+B)\cos{x}+(Cx+d)\sin{x} $$
    We need to derive this, but lets do it cleanly:

    Note the grouping by alphabetical order and trig function
    $$ y_{p2}\prime = (A+Cx+D)\cos{x}+(-Ax-B+C)\sin{x} $$
    $$ y_{p2}\prime\prime = (-Ax-B+2C)\cos{x}+(-2A-Cx-D)\sin{x} $$
    
    Plugging it into the equation (Grouping by trig function):
    
    $x\cos{x}$ terms $\rightarrow$ $((-A)-5(C)+4(A))x\cos{x} $ $\rightarrow$ $3A-5C=0$
    
    $\cos{x}$ terms $\rightarrow$ $((-B+2C)-5(A+D)+4(B))\cos{x} $ $\rightarrow$ $-5A+3B+2C=0$
    
    $x\sin{x}$ terms $\rightarrow$ $((-C)-5(-A)+4(C))x\sin{x}$ $\rightarrow$ $3C+5A=0$

    $\sin{x}$ terms $\rightarrow$ $((-2A-D)-5(-B+C)+4(D))\sin{x}$ $\rightarrow$ $-2A+5B-5C+3D=0$
