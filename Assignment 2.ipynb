# Question 1
Show that $y_1(x)=e^x$ is a solution of the equation
$$ y^{\prime\prime}-4y^{\prime}+3y = 0 $$

Use the Reduction Method to find a second, linearly independent, solution.

## Answer
Assuming a solution of the form $y(x)=e^{mx}$.
$$ (m^2 - 4m + 3)e^{mx} = 0 $$
Since $e^{mx}$ is never 0 we will only consider $m^2 - 4m +3=0$.

$$ (m-1)(m-3)=0 $$
Therefore $ m_1 = 1, m_2 = 3 $.

Using $m_1$, our solution for $y_1(x)=e^{x}$, **thus is it a solution of the equation**.

Using Reduction to find a second, linearly independent solution:
$$ y^{\prime\prime}-4y^{\prime}+3y=0, y_1=e^x $$
Since $y_2$ must be linearly independent, it must be of the form $y_2=u(x)y_1(x)$.
$$ y_2=u(x)y_1(x) $$
$$ y_2^{\prime}= u^{\prime}(x)y_1(x)+u(x)y_1^{\prime}(x) $$
$$ y_2^{\prime\prime}= u^{\prime\prime}(x)y_1(x) + u^{\prime}(x)y_1^{\prime}(x) + u^{\prime}(x)y_1^{\prime}(x)+u(x)y^{\prime\prime}(x) $$

**Note**: We will simpily state $u, y_1, u^{\prime}$ etc from now on for simplicity.

Substituting into the original equation and grouping by $u, u^{\prime}$, and $u^{\prime\prime}$:
$$ u^{\prime\prime}y_1 + u^{\prime}(2y^{\prime}-4y_1) + u(y^{\prime\prime}-4y_1^{\prime}+3) = 0 $$
Noting here that as previously stated, $y^{\prime\prime}-4y_1^{\prime}+3 = 0$.

We'll set $v = u^{\prime}$ and $v^{\prime}=u^{\prime\prime}$.
$$ v^{\prime}y_1 = -v(2y_1^{\prime}-4y_1) $$
$$ \frac{v^{\prime}}{v} = \frac{ -(2y_1^\prime - 4y_1) }{ y_1 } = \frac{ -2y_1^\prime }{ y_1 } + 4 $$

Solving as a seperable equation:
$$ \ln{v} = -2 \int{ \frac{ y_1^\prime }{ y_1 } }dx + 4\int{}dx = -2\ln{y_1}+4x $$
Solving for $v$:
$$ v = e^{-2x+4x} = e^{2x} $$

Getting back to $u$:
$$ u = \int{v}dx = \int{e^{2x}}dx = \frac{e^{2x}}{2} $$

Recalling that $y_2(x)=u(x)y_1(x)$:
$$ y_2(x) = \frac{e^{2x}}{2}e^{x} = \frac{1}{2}e^{3x} $$

**Which is our second solution.**
