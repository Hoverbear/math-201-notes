# Question 1
Show that $y_1(x)=e^x$ is a solution of the equation
$$ y^{\prime\prime}-4y^{\prime}+3y = 0 $$

Use the Reduction Method to find a second, linearly independent, solution.

## Answer
Assuming a solution of the form $y(x)=e^{mx}$.
$$ (m^2 - 4m + 3)e^{mx} = 0 $$
Since $e^{mx}$ is never 0 we will only consider $m^2 - 4m +3=0$.

$$ (m-1)(m-3)=0 $$
Therefore $ m_1 = 1, m_2 = 3 $.

Using $m_1$, our solution for $y_1(x)=e^{x}$, **thus is it a solution of the equation**.

Using Reduction to find a second, linearly independent solution:
$$ y^{\prime\prime}-4y^{\prime}+3y=0, y_1=e^x $$
Since $y_2$ must be linearly independent, it must be of the form $y_2=u(x)y_1(x)$.
$$ y_2=u(x)y_1(x) $$
$$ y_2^{\prime}= u^{\prime}(x)y_1(x)+u(x)y_1^{\prime}(x) $$
$$ y_2^{\prime\prime}= u^{\prime\prime}(x)y_1(x) + u^{\prime}(x)y_1^{\prime}(x) + u^{\prime}(x)y_1^{\prime}(x)+u(x)y^{\prime\prime}(x) $$

**Note**: We will simpily state $u, y_1, u^{\prime}$ etc from now on for simplicity.

Substituting into the original equation and grouping by $u, u^{\prime}$, and $u^{\prime\prime}$:
$$ u^{\prime\prime}y_1 + u^{\prime}(2y^{\prime}-4y_1) + u(y^{\prime\prime}-4y_1^{\prime}+3) = 0 $$
Noting here that as previously stated, $y^{\prime\prime}-4y_1^{\prime}+3 = 0$.

We'll set $v = u^{\prime}$ and $v^{\prime}=u^{\prime\prime}$.
$$ v^{\prime}y_1 = -v(2y_1^{\prime}-4y_1) $$
$$ \frac{v^{\prime}}{v} = \frac{ -(2y_1^\prime - 4y_1) }{ y_1 } = \frac{ -2y_1^\prime }{ y_1 } + 4 $$

Solving as a seperable equation:
$$ \ln{v} = -2 \int{ \frac{ y_1^\prime }{ y_1 } }dx + 4\int{}dx = -2\ln{y_1}+4x $$
Solving for $v$:
$$ v = e^{-2x+4x} = e^{2x} $$

Getting back to $u$:
$$ u = \int{v}dx = \int{e^{2x}}dx = \frac{e^{2x}}{2} $$

Recalling that $y_2(x)=u(x)y_1(x)$:
$$ y_2(x) = \frac{e^{2x}}{2}e^{x} = \frac{1}{2}e^{3x} $$

**Which is our second solution.**

# Question 4
Solve the equation
$$ y^{\prime\prime\prime}+y^{\prime} = \tan{x} $$

### Answer
It makes sense in this problem to do a substitution:
$$ y^\prime = u $$
$$ \therefore u^{\prime\prime}+u = \tan{x} $$

**Finding $u_c$:**
$$ m^2 + 1 = 0 $$
$$ m = \pm i $$

Therefore $ u_c = c_1\sin{x}+c_2\cos{x} $.

**Finding $u_p$:**
$$ W(\cos{x},\sin{x}) = \begin{vmatrix} \cos{x} & \sin{x} \\\ -\sin{x} & \cos{x} \end{vmatrix} = \cos^2{x} + \sin^2{x} = 1 $$
$$ W_1 = \begin{vmatrix} 0 & \sin{x} \\\ \tan{x} & \cos{x} \end{vmatrix} = -\sin{x}\tan{x} $$
$$ W_2 = \begin{vmatrix} \cos{x} & 0 \\\ -\sin{x} & \tan{x} \end{vmatrix} = \cos{x}\tan{x} $$

$$ c_1^\prime = \frac{W_1}{W} = -\sin{x}\tan{x} $$
$$ c_2^\prime = \frac{W_2}{W} = \sin{x} $$

$$ c_1 = \int{c_1^\prime}dx = \log{ \cos{ \frac{x}{2} } - \sin{ \frac{x}{2} } } - \log{ \cos{ \frac{x}{2} } + \sin{ \frac{x}{2} } } + \sin{x} $$
$$ c_2 = \int{c_2^\prime}dx = -\cos{x} $$

Therefore our $u_p$ simplified is:
$$ u_p = \cos{x}\log{ \frac{ \cos{ \frac{x}{2} } - \sin{ \frac{x}{2} } }{ \cos{ \frac{x}{2} } + \sin{ \frac{x}{2} } } } $$

**Finding $y$:**

$$ y^\prime = u = c_1\sin{x} + c_2\cos{x} + \cos{x}\log{ \frac{ \cos{ \frac{x}{2} } - \sin{ \frac{x}{2} } }{ \cos{ \frac{x}{2} } + \sin{ \frac{x}{2} } } } $$

# Question 5
Solve the initial value problem
$$ x^2 y^{\prime\prime} + xy^{\prime} + y = 0, y(1)=1, y^{\prime}(1)=2 $$

### Answer
Using Chachy-Euler.

Assuming $y(x)$ is of the form:
$$ y(x) = x^m $$
$$ y^\prime(x) = mx^{m-1} $$
$$ y^{\prime\prime}(x) = m(m-1)x^{m-2} $$

Substituting back:
$$ x^2m(m-1)x^{m-2} + xmx^{m-1} + x^m = 0 $$
$$ m(m-1)+m+1 = 0 $$
$$ (m^2 + 1) = 0 $$
$$ m_{1,2} = \pm i $$

Using the general solution for complex conjugates:
$$ y(x)=c_1x^\alpha \cos{(\beta\ln{x})}+c_2x^\alpha \sin{(\beta \ln{x})} $$

Gives us our **solution**:
$$ y = c_1\cos{(\ln{x})} + c_2\sin{(\ln{x})} $$
We also need it's derivative for the IVP.
$$ y^\prime = \frac{-c_1\sin{(\ln{1})}}{x} + \frac{c_2\cos{(\ln{x})}}{x} $$

**Solving the IVP:**
$$ y(1) = 1 = c_1\cos{(\ln{1})} + c_2 \sin{(\ln{1})} $$
$$ c_1 = 1 $$

$$ y^\prime (1) = -c_1\sin{(\ln{1})} + c_2\cos{(\ln{1})} $$
$$ c_2 = 1 $$
