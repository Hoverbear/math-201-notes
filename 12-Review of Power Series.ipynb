A power series in $(x-a)$ is an infinite series of the form
$$ \sum_{n=0}^\infty{c_n(x-a)^n} = c_0 + c_1(x-a) + c_2(x-a)^2 +... $$
Such a power series is **centered** at $a$.

For example:
$$ \sum_{n=0}^\infty{(x+3)^n}, a = 3 $$

Another:
$$ \sum_{n=0}^\infty{3^nx^n} = 1+3x+9x^2+... $$
Centered at $0$, so "Power Series in $x$".

## Facts
A power series is **convergent** at some specified value of $x$ if it's sequence of partial sums $ \\{S_N(x)\\}^\infty_{N=1}, S_N(x)=\sum^N_{n=0}{(x-a)^n} $ is convergent. 

Meaning that $ \lim_{N\rightarrow\infty}{ \sum^N_{n=0}{c_n(x-a)^n} } $ exists.

If this limit does not exist, then the series is called **divergent**.

### Interval of Convergence
Every power series has an interval of convergence, which is the set of all real numbers $x$ for which the series converges.

### Radius of Convergence
The radius $R$ of convergence of a power series is called **Radius of convergence**. If $R \gt0$ then the power series converges for $|x-a| \lt R$.

So, it converges at $ a-R \lt x \lt a+R $.

And diverges for $ |x-a | \gt R $.

If the series converges at $a$ then $R=0$.

### Absolute Convergence
Within it's interval of convergence, the power series is **absolutely convergent**, meaning that it's convergent in absolute value.
$$ \sum_{n=0}^\infty{ |c_n(x-a)^n| } $$
This is convergent!

## How do we know if the series is convergent or not?
### Ratio Test
Suppose $c_n\neq0$ for all $n=0,1,2,...$ and that 

$$\lim_{n\rightarrow\infty}{ \frac{c_{n+1}(x-a)^{n+1} }{ c_n(x-a)^n } } = |x-a| \lim_{n\rightarrow\infty}{ | \frac{C_{n+1}}{C_n} | } = L $$

If $L \lt 1$, the series converges absolutely.

If $L \gt 1$, the series diverges.

If $L = 1 $, the test is inconclusive.

The test is always inconclusive at $a-R, a+R$.

### Power Series as Functions
In it's interval of convergence every power series is a continuous, integrable, differentiable function.

$$ y(x)=\sum_{n=0}^\infty{ c_nx^n } = c_0+c_1x+c_2x^2+... $$
Lets differentiate!
$$ y\prime(x) = c_1 + 2c_2x + 3c_3x^2 + ... $$
$$ y\prime(x) = \sum_{n=1}^{\infty}{ c_nnx^{n-1} } $$

Again!
$$ y\prime\prime(x) = 2c_2+6c_3x+12c_4x^2 +... $$
$$ y\prime\prime(x) = \sum_{n=2}^{\infty}{ c_nn(n-1)x^{n-2} } $$

**Be very careful with the indices (The bottom of sums!)**

## Identity Property
If $\sum_{n=0}^{\infty}{ c_n(x-a)^n } = 0, R\gt0 $

For all $x$ in some open interval in $(a-R, a+R)$ then $c_n=0$ for all $n$

## Analytic Functions 
(Leaving power series for a moment and talking in general)

A function is called **analytic at a point $a$** if it can be represented as a power series in $x-a$ with a positive or infinite radius.

## Taylor Series for F
$$ \sum_{n=0}^{\infty}{ \frac{f^{(n)}(a)}{n!} (x-a)^n } = f(a)+\frac{f\prime(a)}{1!}(x-a) + \frac{f\prime\prime(a)}{2!}(x-a)^2 + ... $$

## Maclaurin Series
$$ \sum_{n=0}^{\infty}{ \frac{f^{(n)}(0)}{n!}x^n }= f(0)+\frac{f\prime(0)}{1!}x + \frac{f\prime\prime(0)}{2!}x + ... $$
