## Definifition
An equation of the form
$$ x\prime = F(t)x+g(t) $$
is called **linear**.

If $g \neq 0$ then the equation is called **linear non-homogenious**.
If $g = 0$ then the equation is called **linear homogenious**.

**Ex.**
$$ x\prime = tx+t^2 $$
$$F(t)=t$$
$$g(t)=t^2$$

Linear non-homogenious.

**Ex.**
$$ x\prime=\frac{x}{t}+\sin{t} $$
$$ t \neq 0 $$
$$ f(t)=\frac{1}{t} $$
$$g(t)=\sin{t}$$

Linear non-homogenious.

**Ex.**
$$ x\prime = (\cos{t})x $$
$$ f(t)=\cos{t}$$
$$g(t)=0$$

Linear Homogenious.

## The Method of Variation of Parameters
**Also called: The Method of Variation of Constants**

### Step 1
Consider the equation
$$ \frac{dX}{dt}=f(t)X $$
and obtain the general solution
$$ X(t)=ce^{\phi(t)} $$
Where $\phi\prime = f $

*We don't need to worry about singular solutions really here. They're included.*

### Step 2
For the **non-homogenious** equation, seek a solution of the form
$$ x(t)=c(t)e^{\phi(t)} $$

### Step 3
Subtitute $x$ and $x\prime$ into the non-homogenious equation and eventually see that you can determine
$$ c\prime(t)=g(t)e^{-\phi(t)} $$

### Step 4
Integrate and obtain
$$ c(t)=\int{g(t)e^{-\phi(t)}} dt $$

If you can't compute this, that's it!
If you can...

### Step 5
Substitute this $c$ into the general solution of the non-homogenious equation.

Eg.
$$ x(t)=(G(t)+k)e^{\phi(t)} $$
$\phi$ is such that
$$ \phi\prime=f $$
$G$ is such that
$$ G\prime=ge^{-\phi} $$
And $k$ is a constant.

### Getting from Step 2 to Step 3
Using the chain rule here:
$$ x\prime(t) = \frac{d}{dt} (c(t)e^{\phi(t)})=c(t)\phi\prime(t)e^{\phi(t)}+c\prime(t)e^{\phi(t)} $$

But since
$$ x(t)=c(t)^{\phi(t)} $$
$$ \phi\prime(t)=f(t) $$

$$ x\prime(t)=f(t)x(t)+c\prime(t)e^{\phi(t)} $$

Comparing this with
$$ x\prime(t)=f(t)x(t)+g(t) $$

This means
$$ c\prime(t)=g(t)e^{-\phi(t)} $$

## Examples
**Ex.**
$$ x\prime=2(t)x+e^{t^2} $$
We can see that:
$$f(t) = 2(t)$$
$$g(t)=e^{t^2}$$

1. $$ X\prime=2tX $$
    Seperable, solve it.
    $$ X(t)=ce^{t^2} $$
2. Check into the non-homogenious equation. A solution in the form
    $$ x(t)=c(t)e^{t^2} $$
    $$ x\prime(t)=c\prime(t)e^{t^2}+2tc(t)e^{t^2} $$
3. Substitute $x$ and $x\prime$ into the non-homogenious equation.
    $$ c\prime(t)e^{t^2}+2tc(t)e^{t^2}=2tc(t)e^{t^2}+e^{t^2} $$
    Note how the second part of the left, and the first part of the right cancel out.
    $$ c\prime(t)e^{t^2}=e^{t^2} $$
    (**If C(t) doesn't cancel out, you did it wrong.**)
    $$ c\prime(t)=1 $$
4. Solve the above equation
    $$ c(t)=t+k $$
    $$ k = constant $$
5. The general solution is
    $$ x(t)=(t+k)e^{t^2} $$
    (From step 2)

    k goes from $\infty$ to $-\infty$

**Ex.**
$$ x\prime=\frac{\sin{t}}{t}-\frac{2x}{t} $$
$$ x(\frac{\pi}{2})=1 $$
$$ t\neq0$$

Starting:
$$ X\prime = -\frac{2X}{t} $$
$$ \frac{X\prime}{X}=-\frac{2}{t} \Rightarrow X(t)=\frac{c}{t^2} $$

Check into the original equation a solution of the form
$$ x(t)=\frac{c(t)}{t^2} $$
$$ x\prime(t)=\frac{c\prime(t)t^2-2tc(t)}{t^4} = \frac{c\prime(t)t-2c(t)}{t^3} $$

Back to the original equation:
$$ \frac{c\prime(t)t-2c(t)}{t^3}=\frac{\sin{t}}{t}-\frac{2^{\frac{c(t)}{t^2}}}{t} $$
$$ \frac{c\prime(t)}{t^2}-\frac{2c(t)}{t^3}=\frac{\sin{t}}{t}-\frac{2c(t)}{t^3} $$

Note how the $c(t)$ cancels.
$$ \frac{c\prime(t)}{t^2}=\frac{\sin{t}}{t} $$
$$ c\prime(t)=t\sin{t} $$

(Now you can integrate by parts which leads us too...)
$$ c(t)=\sin{t}-t\cos{t}+k $$

So the general solution of the non-homogenious equation
$$ x(t)=\frac{\sin{t}-t\cos{t}+k}{t^2} $$

Now, to obtain the solution of the initial value problem...
$$ 1=x(\frac{\pi}{2}) = \frac{\sin{\frac{\pi}{2}}-\frac{\pi}{2}\cos{\frac{\pi}{2}}+k}{\frac{\pi^2}{4}} $$
$$ 1=\frac{1+k}{\frac{\pi^2}{4}} \Rightarrow k=\frac{\pi^2}{4}-1 $$

The solution of the initial value problem is $\phi$:
$$ \phi(t)=\frac{4\sin{t}-4t\cos{t}+\pi^2-4}{4t^2} $$

$$ \phi:(0,\infty)\rightarrow \mathbb{R} $$
We choose that because of the initial value of $x(\frac{\pi}{2})$

## Applications

### Ex. A falling object

We have:

* 5 kg falling box.
* $$ F_{resistance}=-\frac{2}{5(t+1)}v $$
    v = velocity,

    t = time in seconds.

Find the folmula of the boxes velocity and solve.
F = Mass * Acceleration
Acceleration = $v\prime$
$$ F=F_{weight}+F_{resistance} $$
$$ F=F_{weight}=mg $$
$$ g = 9.8 $$
$$ w = 5 $$

Equation becomes 
$$ 5v\prime = 5g - \frac{2}{5(t+1}v $$
$$ v\prime = 9.8 - 0.08 \frac{v}{t+1} $$

This is a linear equation!
Solving...

$$ V\prime=-0.08\frac{V}{t+1} \Rightarrow V(t)=c(t+1)^{-0.08} $$
Check into the non-homogenious equation a solution of the form
$$ v(t)=c(t)(t+1)^{-0.08} $$
Obtain:
$$ c\prime(t)=9.8(t+1)^{0.08} $$
$$ c(t)=\frac{9.8}{1.08}(t+1)+k $$
Which leads to the general solution:
$$ v(t)=9.074(t+1)+k(t+1)^{0.08} $$

It is safe to assume $v(0)=0$ to determine the k.
$$ 0=v(0)=9.074+k*1^{-0.08} $$
This means
$$ k = -9.074 $$
$$ \phi=9.074(t+1-(t+1)^{-0.08}) $$
